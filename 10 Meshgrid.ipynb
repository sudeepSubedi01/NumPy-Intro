```python
import numpy as np
import matplotlib.pyplot as plt
```

# Meshgrid
- numpy.meshgrid is a function that generates coordinate grids from two or more one-dimensional arrays.

# Simple Meshgrid

```python
x = np.linspace(0,100,100)
print(x.shape)
```

```python
x1,x2 = np.meshgrid(x,x)
print(x1.shape)
print(x2.shape)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x1,x2)
```

```python
x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)

x,y = np.meshgrid(x,y)

plt.scatter(x,y)
plt.show()
```

# 3d

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5,5,10)
y = np.linspace(-5,5,10)

x,y = np.meshgrid(x,y)

z = x**2 + y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
plt.show()
```
